```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Open the file in binary mode
path = "FFT2.bin"
filename, ext = path.split(".", 1)
with open(path, 'rb') as f:
    # Read the data into a NumPy array
    array = np.fromfile(f, dtype=np.csingle)  # Change dtype according to your data
    f.close()
```

```python
L = 64
L_half = L//2 +1
bit_reversal_index = [0, 32, 16, 48, 8, 40, 24, 56, 4, 36, 20, 52, 12, 44, 28, 60, 2, 34, 18, 50, 10, 42, 26, 58, 6, 38, 22, 54, 14, 46, 30, 62, 1, 33, 17, 49, 9, 41, 25, 57, 5, 37, 21, 53, 13, 45, 29, 61, 3, 35, 19, 51, 11, 43, 27, 59, 7, 39, 23, 55, 15, 47, 31, 63]
Fs = 1000
```

```python
P2 = abs(array/L)
```

```python
array2 = array.copy()
```

```python
for i,j in enumerate(bit_reversal_index):
    array2[i] = array[j] 
```

```python
P2 = abs(array2/L)
P1 = P2[0:L_half]
P1[1:-2] = 2*P1[1:-2]
```

```python
frequency = np.zeros(33)
for i, item in enumerate(frequency):
    frequency[i] = i*Fs/L
```

```python
plt.plot(frequency, P1)
plt.title("Onesided FFT with transform Length 64 of data: "+ filename + ".bin", wrap=True)
plt.savefig(filename + ".png") 
```
